***Q2) What is the use of pandas, numpy, seaborn and matplotlib? Explain with examples***

Ans:

**pandas:** Pandas is a Python library for data manipulation and analysis. It provides high-performance data structures like Series and DataFrames, which are similar to Excel spreadsheets. With Pandas allows to easily load, clean, filter, transform, and analyze data. It offers various functions for handling missing values, grouping data, merging datasets, and is essential for tabular data.

**numpy:** NumPy is a package for scientific computing in Python that provides a high-performance multidimensional array object (ndarray) and for mathematical operations on arrays. NumPy is optimized for numerical computations and is used in various scientific fields including machine learning, data analysis, and data engineering.

**seaborn:** Seaborn is a Python data visualization library built with it's dependency on Matplotlib. It provides a interface for creating statistical visualizations. Seaborn offers functions for creating various plot types including histograms, scatter plots, box plots, and heatmaps, while also providing built-in themes and styles to customize the appearance of visualizations.

**matplotlib:** Matplotlib is a fundamental plotting library in Python which provides a low-level interface for creating a wide range of static visualization. Matplotlib offers a framework for customizing plots, axis labels, titles, and colors.
